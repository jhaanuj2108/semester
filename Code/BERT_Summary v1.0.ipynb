```
import numpy as np 
import pandas as pd 
import nltk
from newspaper import Config, Article, Source
config = Config()
config.MAX_SUMMARY_SENT = 3
nltk.download('punkt')
import re
from summarizer import Summarizer
bert_model = Summarizer()
```

```
newsdata = pd.read_csv('/content/drive/My Drive/3rd Sem/newsdata.csv')
```

```
newsdata.head()
```

```
data = newsdata[['News_ID','URL']]
```

```
import csv
with open('./summarydata.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["News_ID", "Newspaper3k", "BERT"])
```

```
def data_gen(data):
    for row in range(7895, len(data)):
        newsid = data.loc[row].News_ID
        url = data.loc[row].URL
        article = Article(url, config = config)
        article.download() #Downloads the link’s HTML content
        try:
            article.parse() #Parse the article
            article.nlp()#  Keyword extraction wrapper
            body = article.text
            body = body.replace("\n", "")
            newspaper3k_summary = article.summary
            yield newsid, body, newspaper3k_summary
        except:
            pass
```

```
a = data_gen(data)
```

```
def summary(body):
    
    bert_summary = ''.join(bert_model(body, num_sentences=2))
    #print(bert_summary)
    
    return bert_summary
```

```
for item in a:
    newsid = item[0]
    body = item[1]
    newspaper3k_summary = item[2]
    bert_summary = summary(body)
    with open('./summarydata.csv', 'a', newline='') as file:
      writer = csv.writer(file)
      writer.writerow([newsid, newspaper3k_summary, bert_summary])

    print("Current:", newsid)
```
