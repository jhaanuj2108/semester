```
import numpy as np 
import pandas as pd 
import nltk
from newspaper import Config, Article, Source
nltk.download('punkt')
import re
from summarizer import TransformerSummarizer
GPT2_model = TransformerSummarizer(transformer_type="GPT2",transformer_model_key="gpt2-medium")
xlnet_model = TransformerSummarizer(transformer_type="XLNet",transformer_model_key="xlnet-base-cased")
```

```
newsdata = pd.read_csv('/content/drive/My Drive/3rd Sem/newsdata.csv')
```

```
newsdata.head()
```

```
data = newsdata[['News_ID','URL']]
```

```
import csv
with open('./summarydata.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["News_ID", "GPT-2", "XLNET"])
```

```
def data_gen(data):
    for row in range(1519,len(data)):
        newsid = data.loc[row].News_ID
        url = data.loc[row].URL
        article = Article(url)
        article.download() #Downloads the link’s HTML content
        try:
            article.parse() #Parse the article
            article.nlp()#  Keyword extraction wrapper
            body = article.text
            body = body.replace("\n", "")
            yield newsid, body
        except:
            pass
```

```
a = data_gen(data)
```

```
def summary(body):

    gpt2_summary = ''.join(GPT2_model(body, num_sentences=2))
    #print(gpt2_summary)

    xlnet_summary = ''.join(xlnet_model(body,  num_sentences=3))
    #print(xlnet_summary)
    
    return gpt2_summary, xlnet_summary
```

```
for item in a:
    newsid = item[0]
    body = item[1]
    gpt2_summary, xlnet_summary = summary(body)
    with open('./summarydata.csv', 'a', newline='') as file:
      writer = csv.writer(file)
      writer.writerow([newsid, gpt2_summary, xlnet_summary])

    print("Current:", newsid)
```
